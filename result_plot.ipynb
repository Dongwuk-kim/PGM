```python
import json
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
with open('train_result/Beta_5_k_5_summary_dic.json') as json_file:
    json_data_1 = json.load(json_file)
with open('train_result/Beta_5_k_10_summary_dic.json') as json_file:
    json_data_2 = json.load(json_file)
with open('train_result/Beta_5_k_15_summary_dic.json') as json_file:
    json_data_3 = json.load(json_file)
with open('train_result/Beta_5_k_20_summary_dic.json') as json_file:
    json_data_4 = json.load(json_file)
with open('train_result/Beta_5_k_25_summary_dic.json') as json_file:
    json_data_5 = json.load(json_file)
```

```python
result_list = [json_data_1, json_data_2,json_data_3,json_data_4,json_data_5]
```

```python
for key in json_data_1 :
    plt.rcParams["figure.figsize"] = (7,4) 
    plt.figure()
    plt.rcParams['axes.grid'] = True 
    plt.title('{}'.format(key))

    
    for k,result_by_k in enumerate(result_list) :
        plt.plot(result_by_k[key], label = "k = {}".format(5*(k+1)))
    plt.legend()
    plt.savefig('{}.png'.format(key), dpi=300)
```

```python
for key in json_data_1 :
    plt.rcParams["figure.figsize"] = (14,4) 
    plt.figure()
    plt.rcParams['axes.grid'] = True 
    plt.title('{}'.format(key))
    plt.plot(json_data_1[key])
```
